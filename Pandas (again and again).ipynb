```python
dic = {'no' : False, 'yes':True}
dataset['voice mail plan'] = dataset['voice mail plan'].map(dic)
```

```python
dataset.info()
```

```python
dic_reverse = {False:'no', True:'yes'}
dataset = dataset.replace({'voice mail plan': dic_reverse})
dataset.head()
```

```python
columns_to_show = ['total day minutes', 'total eve minutes', 
                   'total night minutes']

dataset.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```

```python
pd.crosstab(dataset['churn'],dataset['area code'])
```

```python
pd.crosstab(dataset['churn'],dataset['voice mail plan'], normalize=True)
```

```python
dataset.pivot_table(['total eve calls','total day calls', 'total night calls'],['area code'], aggfunc='mean')
```

```python
total_calls = dataset['total eve calls'] + dataset['total night calls'] + \
              dataset['total day calls'] + dataset['total intl calls']
```

```python
dataset.insert(loc=dataset.shape[1],column='total calls',value=total_calls)
```

```python
dataset.head()
```

```python
dataset['total charge'] = dataset['total day charge'] + dataset['total eve charge'] + \
                     dataset['total night charge'] + dataset['total intl charge']
dataset.head()
```

```python
dataset.drop(['total calls','total charge'],axis=1,inplace=True)
dataset.drop([1,2]).head()
```

```python
pd.crosstab(dataset['churn'],dataset['international plan'],margins=True)
```
